```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
X = np.array([
    [150, 50],  # Person A
    [160, 60],  # Person B
    [165, 65],  # Person C
    [170, 70],  # Person D
    [180, 80]   # Person E
])
```

```python
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

```python
pca=PCA(n_components=2)
X_pca=pca.fit_transform(X_scaled)
```

```python
print("Eigenvalues (Variance explained):", pca.explained_variance_)
print("Explained variance ratio:", pca.explained_variance_ratio_)
print("Principal Components (Eigenvectors):\n", pca.components_)
```

```python
# Step 4: Plot the original data
plt.figure(figsize=(8,6))
plt.scatter(X_scaled[:,0], X_scaled[:,1], color='blue', label='Original Data')
```

```python

# Step 4: Plot the original data
plt.figure(figsize=(8,6))
plt.scatter(X_scaled[:,0], X_scaled[:,1], color='blue', label='Original Data')
# Draw PC1 and PC2
mean = np.mean(X_scaled, axis=0)
for i, length in enumerate(pca.explained_variance_):
    vector = pca.components_[i] * np.sqrt(length) * 2  # scale for visualization
    plt.arrow(mean[0], mean[1], vector[0], vector[1],
              head_width=0.1, head_length=0.1, color='red' if i==0 else 'green', linewidth=2)
    plt.text(mean[0]+vector[0], mean[1]+vector[1], f'PC{i+1}', color='red' if i==0 else 'green', fontsize=12)

plt.xlabel('Height (standardized)')
plt.ylabel('Weight (standardized)')
plt.title('PCA Visualization: Tiny Dataset')
plt.grid(True)
plt.legend()
plt.axis('equal')
plt.show()
```

```python

# Step 5: Plot data projected onto PC1
plt.figure(figsize=(8,4))
plt.scatter(X_pca[:,0], np.zeros_like(X_pca[:,0]), color='purple', label='Projected onto PC1')
for i, txt in enumerate(['A','B','C','D','E']):
    plt.text(X_pca[i,0], 0.02, txt, fontsize=12, ha='center')
plt.xlabel('PC1 (1D Projection)')
plt.title('Data Projected onto PC1')
plt.grid(True)
plt.yticks([])
plt.legend()
plt.show()
```
